```python
car_data=df[['Year',  'Selling_Price',  'Present_Price',  'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission',  'Owner']]
car_data.head()
car_data['current_year'] = 2020
car_data['no_years'] = car_data['current_year'] - car_data['Year']

car_data.drop(['Year'], axis = 1, inplace = True)
car_data.drop(['current_year'], axis = 1, inplace = True)
car_data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
print(sns.__version__)
sns.pairplot(car_data)
plt.show()
```

```python
X = car_data.iloc[:,1:]
y = car_data.iloc[:,0]
# print(X['Owner'].unique())
print(X.head())
print(y.head())
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
print(model.feature_importances_)
```
